```python
concat_df = pd.concat([snp_data,bnd_data,sret_data,bito_data,upp_data],
            axis=1, keys=['SNP','BND','SRET','BITO','UPP'])
concat_df
```

```python
daily_returns=concat_df.pct_change().dropna()
daily_returns
```

```python
cumulative_returns = (1 + daily_returns).cumprod()
cumulative_returns
```

```python
standard_deviation = daily_returns.std().sort_values()
standard_deviation
```

```python
trading_days = 252
annualized_standard_deviation = standard_deviation * np.sqrt(trading_days)
annualized_standard_deviation
```

```python
annual_average_returns = daily_returns.mean() * trading_days
annual_average_returns.sort_values()
```

```python
sharpe_ratios = annual_average_returns / annualized_standard_deviation
sharpe_ratios.sort_values()
```
